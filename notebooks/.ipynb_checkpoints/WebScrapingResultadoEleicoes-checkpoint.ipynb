### Coleta e tratamento inicial dos dados

```python
# urlopen:
from urllib.request import urlopen
# BeautifulSoup: 
from bs4 import BeautifulSoup
```

```python
url = "https://pt.wikipedia.org/wiki/Elei%C3%A7%C3%A3o_presidencial_no_Brasil_em_2018"
# urlopen faz o quê????
html = urlopen(url)
#type(html)
```

```python
soup = BeautifulSoup(html, "lxml")
#type(soup)
```

```python
#title = soup.title
#print(title)
```

```python
tabelas = soup.find_all("table", {"class":"wikitable", "width":"70%"})
#print(type(tabelas))
#print(len(tabelas))
```

```python
rows = tabelas[0].find_all("tr")
```

```python
lista_dados = []
for row in rows:
    row_td = row.find_all("td")
    dado = [row.text.strip() for row in row_td]
    lista_dados.append(dado)
print(lista_dados)
```

#### Criação dos dataframes e manipulação dos dados

```python
import pandas as pd
```

```python
df_dados_original = pd.DataFrame(lista_dados)
df_dados_original.head(20)
```

```python
#remover valores quando todos os registros no eixo x (axis=0) estiverem sem valor (how="all")
df_dados_sem_missing = df_dados_original.dropna(axis=0, how="all")
#remover valores quando todos os registros no eixo y (axis=1) estiverem sem valor (how="all")
df_dados_sem_missing = df_dados_sem_missing.dropna(axis=1, how="all")
```

```python
df_dados_sem_missing.head(50)
```

```python
#queremos somente os candidatos
df_somente_candidatos = df_dados_sem_missing[0:-5]
```

```python
df_somente_candidatos.head(20)
```

```python
#refazendo os índices
df_final = df_somente_candidatos.reset_index(drop=True)
```

```python
df_final.head(20)
```

```python
heads = tabelas[0].find_all("th")
#print(heads[:10])
```

```python
lista_heads = []
for head in heads:
    head_th = head.find(text=True)
    #cabecalho = [head.text for head in head_th]
    lista_heads.append(head_th.strip())
print(lista_heads)
```

```python
df_heads = pd.DataFrame(lista_heads)
#df_heads.head(50)
```

```python
df_heads = df_heads[0:-16]
```

```python
df_heads.head(50)
```

```python
#excluindo linhas
df_heads = df_heads.drop([2,3,4], axis=0)
df_heads = df_heads.reset_index(drop=True)
df_heads.head(50)
```

```python
#renomear linhas
df_heads.iloc[2,0] = "TotalTurnoI"
df_heads.iloc[3,0] = "PercentualTurnoI"
df_heads.iloc[4,0] = "TotalTurnoII"
df_heads.iloc[5,0] = "PercentualTurnoII"
df_heads.head(50)
```

```python
#transformando linhas em colunas
df_heads = df_heads.T
df_heads.head(50)
```

```python
header = df_heads.iloc[0]
#type(header)
df_completo = df_final.rename(columns = header)
```

```python
df_completo.head(50)
```

```python
df_completo.dtypes
```

```python
#convertendo votação para número, eliminando os espaços na string para evitar erro na conversão
df_completo['TotalTurnoI'] = df_completo.TotalTurnoI.str.replace(' ', '').astype(float)
df_completo['TotalTurnoII'] = df_completo.TotalTurnoII.str.replace(' ', '').astype(float)
```

```python
df_completo.head(20)
```
