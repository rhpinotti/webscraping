```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

### Coleta e limpeza dos dados

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, "lxml")
type(soup)
```

```python
title = soup.title
print(title)
```

```python
text = soup.get_text()
print(text)
```

```python
soup.find_all("a")
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all("tr")
print(rows[:10])
```

```python
lista_linhas = []
for row in rows:
    row_td = row.find_all("td")
    linha = [row.text for row in row_td]
    lista_linhas.append(linha)
df_linhas = pd.DataFrame(lista_linhas)
```

```python
df_linhas.head(10)
```

```python
colunas = soup.find_all("th")
print(colunas[:10])
```

```python
colunas_str = str(colunas)
colunas_str = BeautifulSoup(colunas_str, "lxml").get_text()
lista_colunas = []
lista_colunas.append(colunas_str)
print(lista_colunas)
```

```python
df_colunas = pd.DataFrame(lista_colunas)[0].str.split(',', expand=True)
df_colunas.head()
```

```python
#concatenando os dataframes
frames = [df_colunas, df_linhas]
df_completo = pd.concat(frames)
```

```python
df_completo.head(10)
```

```python
#deixando a primeira linha como cabeçalho e removendo dos registros
df_final = df_completo.rename(columns=df_completo.iloc[0]).drop(df_completo.index[0])
df_final.head()
```

```python
#limpar valores missing
df_limpa = df_final.dropna(axis=0, how="any")
df_limpa.head()
```

```python
df_limpa.rename(columns={"[Place": "Place"}, inplace=True)
df_limpa.rename(columns={" Team]": "Team"}, inplace=True)
df_limpa.head()
```

### Análise dos dados e visualização

```python
#convertendo a coluna 'Chip Time' para minutos, permitindo calcular a média do tempo de chegada (em minutos)
time_list = df_limpa[" Chip Time"].tolist()
time_mins = []
for i in time_list:
    h, m, s = i.split(':')
    math = (int(h) * 3600 + int(m) * 60 + int(s))/60
    time_mins.append(math)
```

```python
#incluir a nova coluna no data frame
df_limpa["Runner_mins"] = time_mins
df_limpa.head()
```

```python
#dados estatísticos dos registros numéricos
df_limpa.describe(include=[np.number])
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 15, 5
```

```python
df_limpa.boxplot(column="Runner_mins")
plt.grid(True, axis='y')
plt.ylabel("Chip Time")
plt.xticks([1], ["Runners"])
```

```python
x = df_limpa["Runner_mins"]
ax = sns.distplot(x, hist=True, kde=True, rug=False, color='m', bins=25, hist_kws={"edgecolor": "black"})
plt.show()
```
